# 2-5 pivot table

pandas의 피봇테이블은 빅데이터에 강하기 때문에 엑셀보다 강하다.

피봇테이블에서 가장 많이 쓴느건 row와 values를 가장 많이 쓰는데,
엑셀에서 row라고 알고 있는게 pandas에서는 index라고 보면된다.

```python
pd.pivot_table(order, index = ?, values=?)
```

```python
pd.pivot_table(order, index = 'state', values='price')
```

```python
pd.pivot_table(order, index = 'address', values='price')
```

기본값은 index의 평균값(mean)이다. 만약에 누적치를 보고싶다면 aggfunc= 'sum' 을 활용하자.

```python
pd.pivot_table(order, index = 'state', values='price', aggfunc='sum')
```

```python
pd.pivot_table(order, index = 'state', values='price', aggfunc=['mean','sum'])
```

위에서 aggfunc에 인자로 들어간 mean과 sum은 판다스에서 제공하는 것들인데.

numpy의 커스터마이즈된 값들도 인자로 넣을 수 있다.

```python
import numpy as np

pd.pivot_table(order, index = 'state', values='price', aggfunc=[np.mean, np.sum])
```

# 2-7 group by

그룹바이는 피봇테이블과 쓰는 목적과 결과가 거의 똑같다.

그리고 서로 상호 호환이 되는데, 기능이 조금 다르다. 알아보자

```python
order['price'].mean() # 평균 결제 금액
```

```python
# 특정 상황에서의 결제금액을 알고싶다. 예를들어 confirm일떄와 cancel일떄의 price

order.groupby('state')['price'].mean()
```

```python
order.groupby('state')['price'].sum()
```

```python
order.groupby(['date', 'state'])['price'].mean()
```

통계연산을 눈으로 확인하고 싶으면 피봇테이블이고

그게 아니라 통계연산을 가지고 그 값을 또 다른 용도로 사용한다고 하면 groupby를 사용한다.

```python
%timeit pd.pivot_table(order, index=['date', 'state'], values='price')
```

```python
%timeit order.groupby(['date', 'state'])['price'].mean()
```

그룹바이가 피봇테이블보다 훨씬 더 빠르다.

결과 출력을 시각화 해야한다면 피봇테이블, 그게 아니라면 그룹바이를 사용하자.

```python
order = pd.concat([order201701, order201702])
order
```

```python
# 이러면 쓰레기같이 보이니까
pd.concat([order, how])
```

```python
# 이렇게 id를 기준으로 옆으로 데이터를 붙여주는게 맞지 ㅎㅎ
pd.concat([order, how], axis='columns')
```

지금은 데이터의 인덱스가 id로 order, how데이터 모두 지정이 되있기 떄문에

# 2-12 merge

```python
url = 'https://bit.ly/dsa-02-user'

user = pd.read_csv(url, index_col = 'id')
user
```

```python
url = 'https://bit.ly/dsa-02-product'

product = pd.read_csv(url, index_col = 'id')
product
```

merge를 할때는 , 공통되는 키를 찾는게 우선이다.

```python
# 유저 데이터에서는 user_id, product_id가 보인다.

order
```

```python
# user와 order를 합쳐보자.

pd.merge(order, user, left_on = 'user_id', right_on = 'id')
```

```python
# 두개의 데이터에 기준이 되는 컬럼의 이름이 같다면 on을 명시하지 않아도 되고,
# 구태여 명시하고 싶으면 on을 써서 명시하자.

user['user_id'] = user.index

pd.merge(order, user, on='user_id')
```

```python
# 걍 해도됌 좀 위험하긴 하지만

pd.merge(order, user)
```

# 2-15 merge 마지막

```python
pd.merge(order, user, on='user_id')
```

```python
order
```

```python
# order와 user간에 동시에 존재하는 데이터가 있겠지만, 그렇지 않은 경우가 있다.
# 여기서 merge에 how 옵션을 집어 넣는다. 둘 다 있는 정보만(inner),
# 그딴거 필요없고 그냥 모든 정보, 비어있는것은 nan(outer),
# 왼쪽에 있는 정보는 모두 살리고 오른쪽은 버리는것 (left), 
# 오른쪽에 있는 정보는 모두 살리고 왼쪽은 버리는것 (right)

pd.merge(order, user, on='user_id', how = 'outer')
```

```python
pd.merge(order, user, on='user_id', how = 'left')
```

```python
pd.merge(order, user, on='user_id', how = 'right')
```
