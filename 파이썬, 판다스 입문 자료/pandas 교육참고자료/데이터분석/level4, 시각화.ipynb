### 4-8 distplot, 분포를 시각화한다. distribution plot

```python
sns.distplot(data['허리둘레'])
```

시각화가 2개가 동시에 된다. bar는 히스토그램 hist, 선은 kde

경계선 주변은 hist를 신뢰하고, 전체적인 추세는 kde를 신뢰해야한다.

```python
sns.distplot(data['허리둘레'], hist = False)
```

```python
sns.distplot(data['허리둘레'], kde = False)
```

```python
data['흡연상태'] # 1: 담배 ㄴ 2: 끊음 3:흡연자
```

```python
one = data[data['흡연상태'] == 1]
two = data[data['흡연상태'] == 2]
three = data[data['흡연상태'] == 3]

sns.distplot(one['허리둘레'], hist=False, label='흡연 안함')
sns.distplot(two['허리둘레'], hist=False, label='흡연 끊음')
sns.distplot(three['허리둘레'], hist=False, label='흡연 하는 중')
```

### 4-9 countplot의 응용, barplot

barplot은 y축을 지정할 수 있지만, y축은 숫자여야하고, 그 평균을 출력해준다.

```python
plt.figure(figsize=(18,3))
sns.barplot(data=data, x='연령대코드(5세단위)', y='수축기혈압')
```

위에서 bar위에 검은색 선이 보이는데, 이는 신뢰구간을 나타낸다.

estimator라는 속성으로 aggfunc같은 효과를 낸다. 누적합을 구하자

```python
import numpy as np

plt.figure(figsize=(18,3))
sns.barplot(data=data, x='연령대코드(5세단위)', y='수축기혈압', estimator=np.sum)
```

```python
plt.figure(figsize=(18,3))
sns.barplot(data=data, y='연령대코드(5세단위)', x='수축기혈압', hue='성별코드')
```

이를 세로로 출력하고 싶다면, orient라는 속성에 'h'를 넣어준다.

```python
plt.figure(figsize=(4,9))
sns.barplot(data=data, y='연령대코드(5세단위)', x='수축기혈압', hue='성별코드', orient='h')
```

### 4-12 pointplot

bar플롯과 비슷하지만, bar는 x축간에 서로의 연관이 없을때 사용하면 되고

point는 그래프 형태로 이어지니까, x축간에 서로 연관이 있을때 사용하면 용이하다.

```python
plt.figure(figsize=(18,3))
sns.pointplot(data=data, x='연령대코드(5세단위)', y='수축기혈압', hue='성별코드')
```

참고, searborn linestyles 속성을 추가해보자.

```python
plt.figure(figsize=(18,3))
sns.pointplot(data=data, x='연령대코드(5세단위)', y='수축기혈압', hue='성별코드', linestyles=[':', '--'])
```

## 4-13 scatterplot, 정보량을 많이 넣어 표현할 수 있는 도구

```python
sns.scatterplot(data=data, x='식전혈당(공복혈당)', y='수축기혈압')
```

식전혈당이 150 언더만 가져오자

size 속성으로 총콜레스테롤이 높을수록 점의 사이즈가 커지고, sizes 속성으로 점의 크기 범위를 지정한다.

```python
plt.figure(figsize=(18,10))

target = data[data['식전혈당(공복혈당)'] < 150]

sns.scatterplot(data=target, x='식전혈당(공복혈당)', y='수축기혈압', hue='허리둘레', size='총콜레스테롤', sizes=(30, 550))
```
