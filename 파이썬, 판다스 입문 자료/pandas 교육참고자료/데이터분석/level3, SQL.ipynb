# SQL

실제 회사에서는 csv형태의 데이터가 많지 않다. csv는 그냥 파일이기 때문에 날라가면 ㅈ된다. 또한 실제 데이터는 엄청 크기때문에 분산처리를 해야한다. csv는 파일이기때문에 분산처리가 불가능하기 때문에 대형 데이터를 저장에 특화된 데이터베이스를 사용한다.

데이터베이스를 조작하는 언어가 SQL임 ㅋㅋ

데이터 분석가가 배워야하는 SQL과, 데이터베이스 관리자가 배워야하는 SQL은 조금 다르다.

SQL의 4대장은 SELECT, INSERT, UPDATE, DELETE가 있는데, 데이터 분석가는 데이터 수정은 하지 않기 떄문에 SELECT만 사용할 줄 알면 된다. 즉, 데이터분석에 필요한 SQL 지식은 적다.

데이터베이스관리자는 SQL 알고리즘의 시간복잡도, 공간복잡도도 신경써야하는데 데이터분석가는 그렇지 않슴 ㅋㅋ

sql언어 종류가 존나 많은데, 그건 설치하는 시간도 오래걸리고 거시기하니께 그냥 sqlite3 라이브러리를 사용한다.

```python
import sqlite3
import pandas as pd
```

```python
connection = sqlite3.connect("e-commerce.sqlite") # sqlite3.connect는 db가 존재하면 불러오지만, 그 이름의 db가 존재하지 않으면 그냥 새로 만들어버린다.
connection
```

```python
connection_test = sqlite3.connect("e-commerce.db") # 일부러 확장자를 다르게 해서
connection_test # 존재하지 않지만 오류를 내지 않고 그냥 새로 만들어버린다 ㅅㅂ
```

```python
query = "SELECT * FROM 'orders'" # 디폴트 쿼리, orders 테이블에서 모든 데이터를 가져와! 묻고 떠블로가!

order = pd.read_sql(query, connection, index_col = 'id')
order
```

```python

query = "SELECT id, date, price FROM 'orders'" # date, price, id만 가져온나!

order = pd.read_sql(query, connection, index_col = 'id')
order
```

```python
query = "SELECT id, date, price FROM 'orders' LIMIT 3" # 3개 행만 가져온나!

order = pd.read_sql(query, connection, index_col = 'id')
order
```

```python
query = "SELECT * FROM 'orders' ORDER BY date" # 시간순으로 정렬해라

order = pd.read_sql(query, connection, index_col = 'id')
order
```

```python
query = "SELECT * FROM 'orders' ORDER BY date DESC" # 역시간순으로 정렬해라
order = pd.read_sql(query, connection, index_col = 'id')
order
```

```python
query = """
    SELECT *
        FROM 'orders' 
        ORDER BY price DESC
        LIMIT 3    
""" # 가격이 가장 비싼 상위 3개만 가져와라
order = pd.read_sql(query, connection, index_col = 'id')
order
```

### 3-10 SQL WHERE, 색인기능을 제공

```python
# SQL문에서는 !=이게 <>임 ㅋㅎ
query = "SELECT * FROM 'orders' WHERE state <> 'confirmed'"

order = pd.read_sql(query, connection, index_col = 'id')
order
```

```python
query = "SELECT * FROM 'orders' WHERE price > 500"

order = pd.read_sql(query, connection, index_col = 'id')
order
```

SQL에서는 BETWEEN을 활용해 범위를 지정해줄 수 있다.

```python
query = "SELECT * FROM 'orders' WHERE price BETWEEN 400 AND 900"

order = pd.read_sql(query, connection, index_col = 'id')
order
```

address가 Seoul이거나 Incheon인 행만 추출

```python
query = "SELECT * FROM 'orders' WHERE address IN ('Seoul', 'Incheon')" # 디폴트 쿼리, orders 테이블에서 모든 데이터를 가져와! 묻고 떠블로가!

order = pd.read_sql(query, connection, index_col = 'id')
order
```

price가 500 이상이고, state가 confirmed인 행 찾기

```python
query = """SELECT *
    FROM 'orders'
    WHERE state = 'confirmed'
    AND price >= 500
    ORDER BY price DESC
    LIMIT 3
    """ # AND, OR을 사용해서 다양한 색인을 추가할 수 있다.

order = pd.read_sql(query, connection, index_col = 'id')
order
```

특정 문자가 포함되있는지를 검사해서 추출하는, SQL문 LIKE '%text%'

```python
query = "SELECT * FROM 'orders' WHERE address LIKE '%d%'" 

order = pd.read_sql(query, connection, index_col = 'id')
order
```

users 테이블의 id와 orders 테이블의 user_id 가 키가 됩니다.

product 테이블의 id와 orders 테이블의 product_id 가 키가 됩니다.

```python
query = """
    SELECT *
        FROM 'orders'
    LEFT JOIN 'users' ON 'orders'.user_id = 'users'.id
"""


pd.read_sql(query, connection)
```

위 데이터는 너무 더러우니까, 필요한 컬럼만 추출해내자

```python
query = """
    SELECT 'orders'.id, 'orders'.date, 'orders'.price, 'users'.name, 'users'.gender
        FROM 'orders'
    LEFT JOIN 'users' ON 'orders'.user_id = 'users'.id
"""

pd.read_sql(query, connection)
```

SQL도 판다스 merge처럼 LEFT, RIGHT, INNER, OUTER 조인 모두 존재하지만,

sqlite에서는 OUTER는 지원하지 않고, RIGHT는 LEFT랑 걍 위치만 바꾸면 되기대문에 잘 쓰이지 않는다.

```python
query = """
    SELECT 'orders'.id, 'orders'.date, 'orders'.price, 'users'.name, 'users'.gender
        FROM 'orders'
    INNER JOIN 'users' ON 'orders'.user_id = 'users'.id
"""

pd.read_sql(query, connection)
```
