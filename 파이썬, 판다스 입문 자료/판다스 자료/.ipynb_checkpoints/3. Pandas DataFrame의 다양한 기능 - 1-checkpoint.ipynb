## 자주 사용하는 수학 연산 함수 소개

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "가로" : [10, 20, 30, 10, 30, 20, 11],
    "세로" : [20, 23, 22, 33, 22, 12, 11],
    "높이" : [50, 40, 20, 50, 20, 30, 40]
})

df
```

sum : 합계

```python
# sum 합계 계산 axis=0
df.sum()
```

```python
# sum 합계 계산 axis=1
df.sum(axis=1)
```

divide : 나누기

```python
df.divide(2)
```

```python
# sum과 함께 응용하자면.. 이런식으로도 사용 가능
df.divide(df.sum(axis=1), axis=0)
```

누적 계산 (누적곱, 누적 최대,최소값)

```python
# cumprod 누적곱

df.cumprod(axis=0)
```

```python
# cumprod 누적곱

df.cumprod(axis=1)
```

```python
# cummax 누적 최댓값

df.cummax()
```

```python
# cummax 누적 최댓값

df.cummax(axis=1)
```

## apply 함수

dataframe에 파이썬 함수를 적용할 수 있다

예를들어, 가로, 세로, 높이를 이용해 부피라는 컬럼을 추가해보자

```python
df
```

```python
def getVolume(row):
    return row['가로'] * row['세로'] * row['높이']

df['부피'] = df.apply(getVolume, axis=1)
df
```

## concat 함수

서로 다른 두 개의 데이터 프레임을 합치는 기능

```python
df
```

```python
df2 = pd.DataFrame({
    "가로" : [10, 20, 30, 10, 30, 20, 11],
    "세로" : [20, 23, 22, 33, 22, 12, 11],
    "높이" : [50, 40, 20, 50, 20, 30, 40]
})

df2
```

```python
pd.concat([df, df2])
```

```python
pd.concat([df, df2], axis=1)
```
