```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = pd.DataFrame([1.01, 0.99, 0.98, 1.02, 1.05], columns=["저변동 수익률"])
b = pd.DataFrame([1.1, 0.9, 0.9, 1.05, 1.1], columns=["고변동 수익률"])
```

```python
a
```

```python
b
```

```python
plt.rc('font', family='Malgun Gothic') 
plt.rc('legend', fontsize=16)

plt.figure(figsize=(12,8))
plt.plot(a, label="저변동")
plt.plot(b, label="고변동")
plt.legend()
plt.show()
```

```python
cumulativeA = a.cumprod()
cumulativeB = b.cumprod()
```

```python
cumulativeA
```

```python
cumulativeB
```

```python
plt.rc('font', family='Malgun Gothic')
plt.rc('legend', fontsize=16)

plt.figure(figsize=(12,8))
plt.plot(cumulativeA, label="저변동")
plt.plot(cumulativeB, label="고변동")
plt.legend()
plt.show()
```
