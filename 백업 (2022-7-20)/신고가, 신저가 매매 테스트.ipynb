```python
import FinanceDataReader as fdr
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pd.set_option('display.max_rows', None)
```

```python
kodex200 = fdr.DataReader("069500", "2000")
```

```python
kodex200 = kodex200['Close']
kodex200
```

```python
kodex200 = pd.DataFrame(kodex200)

window = 20
kodex200["max20"] = kodex200['Close'].rolling(window=window).max()
kodex200["min20"] = kodex200['Close'].rolling(window=window).min()

kodex200 = kodex200.loc["2002-11-08":]
kodex200
```

 :    <= 20  (   .)

 :   >= 20  (   .)

```python
buy = kodex200['Close'] <= kodex200['min20']
sell = kodex200['Close'] >= kodex200['max20']
```

```python
sellPrice= kodex200['Close'].loc[sell]
sellPrice.name = "sell"
```

```python
buyPrice = kodex200['Close'].loc[buy]
buyPrice.name = "buy"
```

```python
a = pd.concat([buyPrice, sellPrice], axis=1)
a
```

```python
isBuy = False
buyPrice = 0
sellPrice = 0
cumReturn = 1
cumReturnSeries = pd.Series()

def getReturn(row):
    global isBuy, buyPrice, sellPrice, cumReturn
    #        
    if not isBuy:
        # row buy 
        # nan pass
        if np.isnan(row['buy']):
            return None
        # nan  isBuy True . buyPrice 
        isBuy = True
        buyPrice = row['buy']
        print(row.name)
        print(": ", buyPrice)
        return row
    
    #     
    # row sell 
    # nan pass
    if np.isnan(row['sell']):
        return None
    # nan  isBuy False , sellPrice .  
    isBuy = False
    sellPrice = row['sell']
    cumReturn *= sellPrice / buyPrice
    cumReturnSeries.loc[row.name] = cumReturn
    print(row.name)
    print(": ", sellPrice)
    print(": ", sellPrice / buyPrice)
    print(": ", cumReturn)
    return row
        
    
b =a.apply(getReturn, axis=1)
```

```python
plt.figure(figsize=(20,8))
cumReturnSeries.plot()
```
