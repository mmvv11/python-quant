```python
import FinanceDataReader as fdr
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pd.set_option('display.max_rows', None)
```

```python
kodex200 = fdr.DataReader("069500", "2000")
```

```python
kodex200 = kodex200['Close']
kodex200
```

```python
kodex200 = pd.DataFrame(kodex200)

window = 20
kodex200["max20"] = kodex200['Close'].rolling(window=window).max()
kodex200["min20"] = kodex200['Close'].rolling(window=window).min()

kodex200 = kodex200.loc["2002-11-08":]
kodex200
```

매수 조건: 오늘 주가  <= 20일 최저가 (주가가 쌀 때 사겠다.)

매도 조건: 오늘 주가 >= 20일 최고가 (주가가 비쌀 때 팔겠다.)

```python
buy = kodex200['Close'] <= kodex200['min20']
sell = kodex200['Close'] >= kodex200['max20']
```

```python
sellPrice= kodex200['Close'].loc[sell]
sellPrice.name = "sell"
```

```python
buyPrice = kodex200['Close'].loc[buy]
buyPrice.name = "buy"
```

```python
a = pd.concat([buyPrice, sellPrice], axis=1)
a
```

```python
isBuy = False
buyPrice = 0
sellPrice = 0
cumReturn = 1
cumReturnSeries = pd.Series()

def getReturn(row):
    global isBuy, buyPrice, sellPrice, cumReturn
    # 매수를 하지 않은 경우    
    if not isBuy:
        # row에 buy를 검사
        # nan이라면 pass
        if np.isnan(row['buy']):
            return None
        # nan이 아니라면 isBuy를 True로 바꾸고. buyPrice 갱신
        isBuy = True
        buyPrice = row['buy']
        print(row.name)
        print("매수가: ", buyPrice)
        return row
    
    # 매수를 하고 보유하고 있는 경우에는
    # row에 sell을 검사
    # nan이라면 pass
    if np.isnan(row['sell']):
        return None
    # nan이 아니라면 isBuy를 False로 바꾸고, sellPrice를 갱신. 누적수익률 갱신
    isBuy = False
    sellPrice = row['sell']
    cumReturn *= sellPrice / buyPrice
    cumReturnSeries.loc[row.name] = cumReturn
    print(row.name)
    print("매도가: ", sellPrice)
    print("매매수익률: ", sellPrice / buyPrice)
    print("누적수익률: ", cumReturn)
    return row
        
    
b =a.apply(getReturn, axis=1)
```

```python
plt.figure(figsize=(20,8))
cumReturnSeries.plot()
```
