```python
import FinanceDataReader as fdr
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#### 함수

```python
def getCloseData(ticker, start, end=None):
    """
    종가 데이터
    ticker: 종목 번호
    start: 시작일
    end: 마지막 날짜
    return: 종목의 종가 데이터
    """
    return fdr.DataReader(ticker, start, end)['Close']

def getDayReturn(closeDataSet):
    """
    개별종목 일별 수익률
    closeDataSet: 종가 데이터
    return: 종가 데이터의 일별 수익률
    """
    return (closeDataSet / closeDataSet.shift(1)).fillna(1)

def getCumulativeReturn(closeDataSet=None):
    """
    개별종목 누적수익률 == 자산흐름
    closeDataSet: 종가 데이터
    return:종가데이터 누적수익률
    """
    return closeDataSet / closeDataSet.iloc[0]

def getPortfolioResult(closeDataSet, weight=None):
    """
    포트폴리오 결과
    closeDataSet: 종가 데이터
    weight: 포트폴리오 개별자산 비중
    return: 포트폴리오 일간수익률, 누적수익률
    """
    # 개별종목 일별 수익률
    dayReturn = getDayReturn(closeDataSet)
    # 개별종목 누적 수익률
    cumulativeReturn = getCumulativeReturn(closeDataSet)
    # 자산별 비중. 기본값: 동일비중
    if not weight:
        weight = [1/len(closeDataSet.columns)] * len(closeDataSet.columns)
        
    # 자산별 비중이 반영된 개별자산 누적 수익률
    weightedCumulativeReturn = weight * cumulativeReturn
    # 자산흐름 정규화: 시기별 각 자산 보유 비율
    normalizedAssetFlow = weightedCumulativeReturn.divide(weightedCumulativeReturn.sum(axis=1), axis=0) # 각 행의 합으로 각 행을 나누기
    # 자산별 비중이 반영된 개별자산 일별 수익률
    weightedDayReturn = normalizedAssetFlow * dayReturn.shift(-1).fillna(0)
    # 포트폴리오 일별 수익률 = 자산별 비중이 반영된 개별자산 일별 수익률의 합
    portfolioDayReturn = weightedDayReturn.sum(axis=1).shift(1).fillna(1)
    # 포트폴리오 누적 수익률
    portfolioCumulativeReturn = portfolioDayReturn.cumprod()
#     portfolioCumulativeReturn = weightedCumulativeReturn.sum(axis=1)
    
    return portfolioDayReturn, portfolioCumulativeReturn

def getEvaluation(dayReturn, cumulativeReturn):
    """
    cagr, mdd, vol
    투자 성과 지표
    """
    # cagr
    cagr = cumulativeReturn.iloc[-1] ** (252/len(cumulativeReturn))
    # mdd
    dd = (cumulativeReturn.cummax() - cumulativeReturn) / cumulativeReturn.cummax() * 100
    mdd= dd.max()
    # vol
    vol = np.std(dayReturn) * np.sqrt(252/len(dayReturn))
    
    return cagr, mdd, vol
```

# 채권 혼합

채권을 혼합하는 이유는 일반적으로 채권은 주식과 가격 움직임이 반대이기 때문입니다.  
서로 반대로 움직이는 종목으로 포트폴리오를 구성하여 한쪽이 하락할 때 그 손실을 줄이는 것이 목적입니다.

```python
# 코스피 ETF 채권 ETF 1:1 혼합

kodex200 = fdr.DataReader("069500", "2015") # kodex 코스피 200 etf
kodex10Bond = fdr.DataReader("152380", "2015") # kodex 국고채 10년물

print(kodex200.head())
print(kodex10Bond.head())
```

```python
# 결측치 가볍게 검사
print(pd.isnull(kodex200).sum())
print(pd.isnull(kodex10Bond).sum())
```

**주식과 채권의 움직임 비교**

```python

plt.figure(figsize=(12, 5))
kodex200['Close'].plot(label="stock")
kodex10Bond['Close'].plot(label="bond")
plt.legend()
plt.show()
```

가격 수준이 다르기 때문에 비교가 쉽지 않습니다.  
정규화를 통해 해결해볼까요?

```python
avgKodex200 = kodex200['Close'].mean() # 평균
stdKodex200 = kodex200['Close'].std() # 표준편차
kodex200['정규화가격'] = (kodex200['Close'] - avgKodex200) / stdKodex200

avgKodex10Bond = kodex10Bond['Close'].mean() # 평균
stdKodex10Bond0 = kodex10Bond['Close'].std() # 표준편차
kodex10Bond['정규화가격'] = (kodex10Bond['Close'] - avgKodex10Bond) / stdKodex10Bond0
```

```python
# 정규화 가격 시각화
plt.figure(figsize=(20, 8))
kodex200['정규화가격'].plot(label="stock")
kodex10Bond['정규화가격'].plot(label="bond")
plt.legend()
plt.show()
```

최근 2022년부터는 금리인상으로 인해 채권과 주식 가격이 커플링되는 현상이 나타났지만, 일반적으로 채권과 주식은 반대로 움직이는 것을 확인했습니다.

### 주식에 분산투자만 했을 때 성과

우선 KODEX200은 코스피 200에 랭크된 종목에 골고루 분산투자하는 ETF입니다.  
이로 인해 분산투자를 했다고 가정하고,
주식 분산 투자 성과를 계산해보겠습니다.

```python
kodex200['일간수익률'] = (kodex200['Close'] / kodex200['Close'].shift(1)).fillna(1)
# kodex200['누적일간수익률'] = kodex200['일간수익률'].cumprod()
kodex200['누적일간수익률'] = kodex200['Close'] / kodex200['Close'].iloc[0]

# cagr
cagr = kodex200['누적일간수익률'].iloc[-1] ** (252/len(kodex200))
# mdd
kodex200["dd"] = (kodex200['누적일간수익률'].cummax() - kodex200['누적일간수익률']) / kodex200['누적일간수익률'].cummax() * 100
mdd = kodex200['dd'].max()
# vol
vol = np.std(kodex200['일간수익률']) * np.sqrt(252/len(kodex200))

print("cagr: ",cagr)
print("mdd: ",mdd)
print("vol: ",vol)
```

```python
plt.figure(figsize=(12, 8))
kodex200['누적일간수익률'].plot(label="stock")
plt.legend()
plt.show()
```

연평균 수익률이 7%이면 나쁘지 않을 수 있지만, MDD가 약 40%입니다.  
견디기 쉽지 않은 수치입니다.

## 채권 ETF와 1:1 혼합할 때 성과

```python
# 데이터 로드
data = pd.DataFrame()
data = pd.concat([data, kodex200['Close'], kodex10Bond['Close']], axis=1)
data.columns = ['kodex200', 'kodex10Bond']
data
```

```python
# 일별 수익률
dayReturn = (data / data.shift(1)).fillna(1)
dayReturn
```

```python
# 자산 흐름 = 누적수익률 ㅇㅇ
assetFlow = data / data.iloc[0]
assetFlow
```

```python
# 자산별 비중. 여기서는 동일 비중으로 가정
portfolioWeight = [1/len(data.columns)] * len(data.columns)
portfolioWeight
```

```python
# 자산별 비중이 반영된 자산 흐름
weightedAssetFlow = portfolioWeight * assetFlow
weightedAssetFlow
```

```python
# 자산 흐름 정규화
normalizedAssetFlow = weightedAssetFlow.divide(weightedAssetFlow.sum(axis=1), axis=0)
normalizedAssetFlow
```

```python
# 포트폴리오 개별자산 순 수익률
individaulNetReturn = normalizedAssetFlow * dayReturn.shift(-1).fillna(0)
individaulNetReturn
```

```python
# 포트폴리오 전체 순수익률
netReturn = individaulNetReturn.sum(axis=1).shift(1).fillna(1)
netReturn
```

```python
# 포트폴리오 누적 수익률
cumulativeReturn = netReturn.cumprod()
cumulativeReturn
```

```python
# cagr
cagr = cumulativeReturn.iloc[-1] ** (252/len(cumulativeReturn))
# mdd
dd = (cumulativeReturn.cummax() - cumulativeReturn) / cumulativeReturn.cummax() * 100
mdd= dd.max()
# vol
vol = np.std(netReturn) * np.sqrt(252/len(netReturn))

print(f"cagr: {cagr}\nmdd: {mdd}\nvol: {vol}")
```

**시각화**

```python
plt.figure(figsize=(20, 15))

# 수익곡선
plt.subplot(2,1,1)
kodex200['누적일간수익률'].plot(label="stock")
cumulativeReturn.plot(label="portfolio")
plt.legend()

# dd 곡선
plt.subplot(2,1,2)
plt.plot(kodex200['dd'] * -1, label="stock drop rate")
plt.plot(dd * -1, label="portfolid drop rate")
plt.legend()

plt.show()
```

코스피200 ETF에만 투자했을 땐 MDD가 38%에 달했지만  
채권 ETF를 혼합했을 땐 MDD가 18%로 약 20%가 감소하여 효과적으로 리스크를 관리했습니다.
