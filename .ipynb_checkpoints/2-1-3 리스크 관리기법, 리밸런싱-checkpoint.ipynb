```python
import FinanceDataReader as fdr
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def getCloseData(ticker, start, end=None):
    """
    종가 데이터
    ticker: 종목 번호
    start: 시작일
    end: 마지막 날짜
    return: 종목의 종가 데이터
    """
    return fdr.DataReader(ticker, start, end)['Close']

def getDayReturn(closeDataSet):
    """
    개별종목 일별 수익률
    closeDataSet: 종가 데이터
    return: 종가 데이터의 일별 수익률
    """
    return (closeDataSet / closeDataSet.shift(1)).fillna(1)

def getCumulativeReturn(closeDataSet=None):
    """
    개별종목 누적수익률 == 자산흐름
    closeDataSet: 종가 데이터
    return:종가데이터 누적수익률
    """
    return closeDataSet / closeDataSet.iloc[0]

def getPortfolioResult(closeDataSet, weight=None):
    """
    포트폴리오 결과
    closeDataSet: 종가 데이터
    weight: 포트폴리오 개별자산 비중
    return: 포트폴리오 일간수익률, 누적수익률
    """
    # 개별종목 일별 수익률
    dayReturn = getDayReturn(closeDataSet)
    # 개별종목 누적 수익률
    cumulativeReturn = getCumulativeReturn(closeDataSet)
    # 자산별 비중. 기본값: 동일비중
    if not weight:
        weight = [1/len(closeDataSet.columns)] * len(closeDataSet.columns)
        
    # 자산별 비중이 반영된 누적 수익률 == 자산흐름
    weightedCumulativeReturn = weight * cumulativeReturn
    # 자산흐름 정규화: 시기별 각 자산 보유 비율
    normalizedAssetFlow = weightedCumulativeReturn.divide(weightedCumulativeReturn.sum(axis=1), axis=0) # 각 행의 합으로 각 행을 나누기
    # 개별자산 가중 수익률
    individaulWeightedReturn = normalizedAssetFlow * dayReturn.shift(-1).fillna(0)
    # 포트폴리오 일별 수익률 = 개별자산 순 수익률의 합
    portfolioDayReturn = individaulWeightedReturn.sum(axis=1).shift(1).fillna(1)
    # 포트폴리오 누적 수익률
    portfolioCumulativeReturn = portfolioDayReturn.cumprod()
    return portfolioDayReturn, portfolioCumulativeReturn

def getEvaluation(dayReturn, cumulativeReturn):
    """
    cagr, mdd, vol
    투자 성과 지표
    """
    # cagr
    cagr = cumulativeReturn.iloc[-1] ** (252/len(cumulativeReturn))
    # mdd
    dd = (cumulativeReturn.cummax() - cumulativeReturn) / cumulativeReturn.cummax() * 100
    mdd= dd.max()
    # vol
    vol = np.std(dayReturn) * np.sqrt(252/len(dayReturn))
    
    return cagr, mdd, vol
```

### 코스피 ETF 채권 ETF 1:1 혼합 && 매월 말 리밸런싱


```python
# 리밸런싱 일자 계산하기

# 주식, 채권 종가 데이터만 가져오기
data = pd.concat([kodex200, kodex10Bond], axis=1)
data.columns = ['kodex200', 'kodex10Bond']
data
```

```python
data['year'] = data.index.year
data['month'] = data.index.month
```

```python
rebalancingDate = data.drop_duplicates(['year', 'month'], keep="last").index
data.drop(['year', 'month'], axis=1, inplace=True) # 쓸모없어진 컬럼 드랍
rebalancingDate
```

```python
# 월별 리밸런싱 비중
# 나중엔 이걸 베이스로 모멘텀 비중으로 해도 되고,,커스텀 가능
portfolioWeight = [0.5, 0.5]
targetWeight = pd.DataFrame([portfolioWeight] * len(rebalancingDate), index=rebalancingDate, columns=data.columns) 
targetWeight
```

```python
unitPortfolioList = [] # 기간별 포트폴리오
previousFinalReturn = 1 # 이전 달 총 수익률, 초기값 1 => 초기 투자금이라고 봐도 되겠네

start = rebalancingDate[0] # 리밸런싱 날짜, 초기값 첫 투자일

for end in rebalancingDate[1:]:
    unitWeight = targetWeight.loc[start] # 당기 리밸런싱 비율
    
    unitPriceData = data.loc[start: end] # 당기 가격 데이터
#     unitReturn = (unitPriceData / unitPriceData.shift(1)).fillna(1) # 당기 개별자산 일간 수익률
    unitCumulativeReturn = unitPriceData / unitPriceData.iloc[0] # 당기 개별자산 누적 수익률

    
    # 당기 포트폴리오 = 이전달 최종 누적수익 * 당기 리밸린성 비율 * 당기 누적수익률 => 자산 흐름에 이전달 최종 누적수익을 곱해서, 이전 흐름이 고려된 자산흐름
    unitPortfolio = previousFinalReturn * unitWeight * unitCumulativeReturn
    unitPortfolioCumulativeReturn = unitPortfolio.sum(axis=1) # 당기 포트폴리오 누적 수익률
    
    unitPortfolio['total'] = unitPortfolioCumulativeReturn
    unitPortfolioList.append(unitPortfolio)
    
    print("당기 누적 수익률")
    print("--당기 자산 흐름을 보여준다.")
    print(unitCumulativeReturn, '\n')
    print("당기 리밸린성 비율 * 당기 누적수익률")
    print("--리밸런싱 비율대로 투자했을 때 자산흐름을 의미한다.")
    print(unitWeight * unitCumulativeReturn, '\n')
    print("전기 최종 누적 수익 = ", previousFinalReturn, '\n')
    print("이전달 최종 누적수익 * 당기 리밸린성 비율 * 당기 누적수익률")
    print("--전달 총 수익률을 곱해줌으로써, 전달 자산 흐름을 반영하여 리밸런싱을 할 수 있다.")
    print((previousFinalReturn * unitWeight) * unitCumulativeReturn, '\n')    
    print("======================\n\n")
    
    previousFinalReturn = unitPortfolioCumulativeReturn.iloc[-1] # 이전 달 최종 누적 수익률 업데이트
    start = end # 리밸런싱 날짜업데이트
```

```python
unitPortfolioList = pd.concat(unitPortfolioList)
unitPortfolioList.head(50)
```

```python
# cagr
cagr = unitPortfolioList['total'].iloc[-1] ** (252/len(unitPortfolioList))
# mdd
unitPortfolioList["dd"] = (unitPortfolioList['total'].cummax() - unitPortfolioList['total']) / unitPortfolioList['total'].cummax() * 100
mdd = unitPortfolioList['dd'].max()
# vol
vol = np.std(unitPortfolioList['total']) * np.sqrt(252/len(unitPortfolioList))

print("cagr: ",cagr)
print("mdd: ",mdd)
print("vol: ",vol)
```

```python
unitPortfolioList['total'].plot()
```

## 리밸런싱 코드

일별수익률을 이용해서 리밸런싱 가중치를 곱해주는 방법

```python
# 샘플 데이터

df = pd.DataFrame(
    {
        "A": [10, 15, 12, 13, 10, 11, 12],
        "B": [10, 10, 8, 13, 12, 12, 12],
        "C": [10, 12, 14, 16, 14, 14, 16],
    },
    index=pd.to_datetime(["2018-01-31", "2018-02-10", "2018-02-20", "2018-02-28", "2018-03-20", "2018-03-29", "2018-04-30",])
)
df
```

```python
# 각 자산별 일별 수익률 계산
dayReturn = (df / df.shift(1)).fillna(1)
dayReturn
```

```python
# 매월 마지막 영업일(리밸린싱 날짜) 계산
df['year'] = df.index.year
df['month'] = df.index.month

rebalancingDate = df.drop_duplicates(['year', 'month'], keep="last").index
df.drop(['year', 'month'], axis=1, inplace=True) # 쓸모없어진 컬럼 드랍
rebalancingDate
```

```python
# 단위 기간(월간) 누적 수익률 계산하기

unitCumulativeReturnList = []

for start, end in zip(rebalancingDate[:-1], rebalancingDate[1:]):
    print(start, end)
    unitPrice = df.loc[start:end] # n기 월말 ~ n+1기 월말 데이터
    unitCumulativeReturn = unitPrice / unitPrice.iloc[0] # n기 월말 ~ n+1기 월말 누적수익률
    unitCumulativeReturnList.append(unitCumulativeReturn)
    
unitCumulativeReturnList
```

```python
# 월간 자산 흐름
monthAssetFlow = pd.concat(unitCumulativeReturnList)
monthAssetFlow
```

```python
# 매월 말 데이터가 겹친다. (2-28, 3-29 등)  
# 매월 말에는 리밸런싱을 하기 때문에 중복되는 행의 마지막 행만 남긴다. 
monthAssetFlow = monthAssetFlow.loc[~monthAssetFlow.index.duplicated(keep='last')]
monthAssetFlow
```

`순수익률` = `자산 비중` * `수익률` 이기 때문에  
월간 자산 흐름 데이터를 통해서 구간별 자산 비중을 계산한다.

```python
portfolioWeight = [0.3, 0.5, 0.2] # 포트폴리오 비중을 곱해준다. a:b:c 해주면 됌

# 월간자산흐름에 포트폴리오 비중을 곱해주면, 매달 말 포트폴리오 비중대로 재조정을 할 수 있고
# 리밸런싱이 들어간 자산흐름을 계산할 수 있다.
rebalancedMonthAssetFlow = monthAssetFlow * portfolioWeight 
rebalancedMonthAssetFlow
```

```python
# 리밸런싱 자산흐름 값을 정규화하여 각 구간별 개별 자산 비중을 계산할 수 있다.
# 쉽게 말하자면, 각 시기별로 어떤 자산을 어느정도(비율) 가지고 있냐 = weight
weight = rebalancedMonthAssetFlow.divide(rebalancedMonthAssetFlow.sum(axis=1), axis=0)
weight
```

```python
# weight는 기간별로 각 자산을 얼마만큼 가지고 있는지 비중을 나타낸다.
# weight와 개별 자산의 수익률을 곱하면 포트폴리오의 수익률을 구할 수 있다.
dayReturn
```

```python
# 포트폴리오 개별자산 순수익률
portfolio = weight * dayReturn.shift(-1).fillna(0)
portfolio
```

```python
# 포트폴리오 전체 순수익률
portfolioRetrun = portfolio.sum(axis=1).shift(1).fillna(1)
portfolioRetrun
```

```python
# 포트폴리오 누적수익률
portfolioRetrun.cumprod()
```

## 리밸런싱 코드
누적수익률을 이용하는 방법

```python
df
```

```python
rebalancingDate
```

```python
# 월별 리밸런싱 비중
# 나중엔 이걸 베이스로 모멘텀 비중으로 해도 되고,,커스텀 가능
targetWeight = pd.DataFrame([portfolioWeight] * len(rebalancingDate), index=rebalancingDate, columns=df.columns) 
targetWeight
```

```python
rebalancingDate[0]
```

```python
# 1. 월별 첫번째 영업일을 기준으로 당기 누적 수익률을 계산한다.
# 2. 월별 리밸런싱 비중과 이전 달 최종 누적 수익률을 곱하면, 이전 달에서 넘어온 총 자산을 이번 달 리밸런싱 비율에 맞게 분배할 수 있다.
# 3. 2에서 구한 (이번달 리밸런싱 비율*이전달 최종 누적 수익)과 당기 누적 수익률을 곱하면, 리밸런싱이 들어간 포트폴리오를 계산할 수 있다.

unitPortfolioList = [] # 기간별 포트폴리오
previousFinalReturn = 1 # 이전 달 총 수익률, 초기값 1 => 초기 투자금이라고 봐도 되겠네

start = rebalancingDate[0] # 리밸런싱 날짜, 초기값 첫 투자일
for end in rebalancingDate[1:]:
    unitWeight = targetWeight.loc[start] # 당기 리밸런싱 비율
    
    unitPriceData = df.loc[start: end] # 당기 가격 데이터
#     unitReturn = (unitPriceData / unitPriceData.shift(1)).fillna(1) # 당기 일간 수익률
    unitCumulativeReturn = unitPriceData / unitPriceData.iloc[0] # 당기 누적 수익률

    
    # 당기 포트폴리오 = 이전달 최종 누적수익 * 당기 리밸린성 비율 * 당기 누적수익률
    unitPortfolio = (previousFinalReturn * unitWeight) * unitCumulativeReturn
    unitPortfolioCumulativeReturn = unitPortfolio.sum(axis=1) # 당기 포트폴리오 누적 수익률
    
    unitPortfolio['total'] = unitPortfolioCumulativeReturn
    unitPortfolioList.append(unitPortfolio)
    
    print("당기 누적 수익률")
    print("--당기 자산 흐름을 보여준다.")
    print(unitCumulativeReturn, '\n')
    print("당기 리밸린성 비율 * 당기 누적수익률")
    print("--리밸런싱 비율대로 투자했을 때 자산흐름을 의미한다.")
    print(unitWeight * unitCumulativeReturn, '\n')
    print("전기 최종 누적 수익 = ", previousFinalReturn, '\n')
    print("이전달 최종 누적수익 * 당기 리밸린성 비율 * 당기 누적수익률")
    print("--전달 총 수익률을 곱해줌으로써, 전달 자산 흐름을 반영하여 리밸런싱을 할 수 있다.")
    print((previousFinalReturn * unitWeight) * unitCumulativeReturn, '\n')    
    print("======================\n\n")
    
    previousFinalReturn = unitPortfolioCumulativeReturn.iloc[-1] # 이전 달 최종 누적 수익률 업데이트
    start = end # 리밸런싱 날짜업데이트
```

```python
# 리밸런싱 날짜엔 row가 2개다. total 값은 같아.
# 앞에꺼는 리밸런싱 전 데이터
# 뒤에꺼는 리밸런싱 후 데이터

portfolio = pd.concat(unitPortfolioList)
portfolio
```

```python
portfolio = portfolio.loc[~portfolio.index.duplicated(keep='last')]
portfolio
```
